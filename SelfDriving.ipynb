```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
image = cv2.imread('/content/drive/MyDrive/images_traffic.jpg')
```

```python
cv2_imshow(image)
```

```python
hsv_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
```

```python
red_lower = np.array([0, 100, 100])
red_upper = np.array([10, 255, 255])
green_lower = np.array([40, 40, 40])
green_upper = np.array([80, 255, 255])
yellow_lower = np.array([20, 100, 100])
yellow_upper = np.array([30, 255, 255])
```

```python
# Create masks to isolate colors
red_mask = cv2.inRange(hsv_image, red_lower, red_upper)
green_mask = cv2.inRange(hsv_image, green_lower, green_upper)
yellow_mask = cv2.inRange(hsv_image, yellow_lower, yellow_upper)

# Count the number of non-zero pixels in each mask
red_pixels = cv2.countNonZero(red_mask)
green_pixels = cv2.countNonZero(green_mask)
yellow_pixels = cv2.countNonZero(yellow_mask)

```

```python
# Count the number of non-zero pixels in each mask
red_pixels = cv2.countNonZero(red_mask)
green_pixels = cv2.countNonZero(green_mask)
yellow_pixels = cv2.countNonZero(yellow_mask)

```

```python
if red_pixels > green_pixels and red_pixels > yellow_pixels:
    print("Stop")
elif green_pixels > red_pixels and green_pixels > yellow_pixels:
    print("GO")
elif yellow_pixels > red_pixels and yellow_pixels > green_pixels:
    print("Ready")
else:
    print("Color undetermined or no traffic light found")

```
